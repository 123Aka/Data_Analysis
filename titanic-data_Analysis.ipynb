```python
# add a new column in the dataset based on the age 

def age_converter(titanic_df):
    for i in range(len(titanic_df['Age'])):
        if titanic_df['Age'][i] <= 10:
            titanic_df['Age'][i] = "kid"
        elif titanic_df['Age'][i] <= 20:
            titanic_df['Age'][i] = "teen"
        elif titanic_df['Age'][i] <= 30:
             titanic_df['Age'][i] = "young"
        elif titanic_df['Age'][i] <= 40:
             titanic_df['Age'][i] = "lower_middle"
        elif titanic_df['Age'][i] <= 50:
            titanic_df['Age'][i] = "upper_middle"
        elif titanic_df['Age'][i] <= 60:
            titanic_df['Age'][i] = "elderly"
        elif titanic_df['Age'][i] > 60 :
            titanic_df['Age'][i] = "old"
        return titanic_df
        
        
        
```

```python
#sort the ages into logical categories
def simplify_ages(df):
    bins = (0, 5, 13, 18, 25, 35, 60, 120)
    group_names = ['Baby', 'Child', 'Teenager', 'Student', 'Young Adult', 'Adult', 'Senior']
    categories = pd.cut(df.Age, bins, labels=group_names)
    df.Age = categories
    return df


simplify_ages(titanic_df)
```
